```python
# 5. How would you catch and handle division by zero error in Python

def divide(a,b):
  try:
    return a/b
  except ZeroDivisionError:
    print("Cannot divide by zero")

divide(10,0)
```

```python
#18. How would you implement a basic logging setup that logs to a file with rotation after 1MB
import logging
from logging.handlers import RotatingFileHandler

logging.basicConfig(level=logging.DEBUG,
                    format='%(asctime)s - %(levelname)s - %(message)s')

logging.handlers.RotatingFileHandler(filename="app.log",
        maxBytes=1 * 512,
        backupCount=5)
for i in range(1000):
  logging.debug(f"This is a debug message {i}")
  logging.info(f"This is an info message {i}")
  logging.warning(f"This is a warning message {i}")
  logging.error(f"This is an error message {i}")
  logging.critical(f"This is a critical message {i}")
```

```python
# 19. Write a program that handles both IndexError and KeyError using a try-except block

def dictstat(objtype,key):
  if type(objtype) == dict:
    try:
       print(objtype[key])
    except KeyError as e:
      print(f"Key not found {e}")

  if type(objtype) == list:
    try:
      print(objtype[key])
    except IndexError as e:
      print(f"Index not found {e}")

dictstat({"name":"Amit", "age":"15", "Class":"10"},"name")
dictstat(["Amit","Sumit","Ankit"],3)

```

```python
# 21. Write a Python program that reads a file and prints the number of occurrences of a specific word?
def readfile(filename,word):
  with open(filename, "r") as file:
      contents = file.read()
  if word in contents:
    print(f"The word '{word}' occurs {contents.count(word)} times")
  else:
    print(f"The word '{word}' does not occur in the file")

readfile('ReadFile.txt', "The")
```
